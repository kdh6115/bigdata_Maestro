```python
import numpy as np
import tensorflow as tf
import pandas as pd
import os
import pathlib
import random
import glob
import matplotlib.pyplot as plt
from skimage.transform import resize
from numpy.random import randn
from matplotlib.image import imread
from keras.preprocessing import image
from PIL import Image
import os, glob
import numpy as np
from sklearn.model_selection import train_test_split
```

```python
img_dir = "C:/Users/ICT01_05/Desktop/3ì°¨/images"
img_dir = pathlib.Path(img_dir)
```

```python
label_names={'product' : 0, 'potg' : 1}
label_key=['product', 'potg']
```

```python
all_img = list(img_dir.glob('*/*.jpg'))
all_img = [str(path) for path in all_img]
random.shuffle(all_img)
all_labels=[label_names[pathlib.Path(path).parent.name] for path in all_img]
```

```python
x=[]
y=[]
for i,f in enumerate(all_img) :
    img = Image.open(f)
    img = img.convert("RGB")
#     img = img.resize((image_w, image_h))
    data = np.asarray(img)
    x.append(data)
    y.append(all_labels)
    if i % 10 == 0:
        print(i, '\n', data)
        
x = np.array(x)
y = np.array(y)
x
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y)
x_train = x_train.astype("float64")/256
# x_test = x_test.astype("float")/256
x_train
```

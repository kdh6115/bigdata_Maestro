```python
# Boolean indexing (True/False)
# 문자열은 대소문자 구분
names = np.array(['Seoul', 'Daejun', 'chungju', 'seoul', 'chungju', 'Daejun', 'Daejun'])
data = np.random.randn(7, 4) # 정규분포에서 출력
print(names == 'Seoul') # 1,0,0,0,0,0,0
print(data) 
print(type(data))
print(data[names == 'Seoul']) # 0번행 #1,0,0,0,0,0,0
print(data[names == 'Seoul', 2:])
print(data[names == 'Seoul', 3])
print(names != 'Seoul')
print(~(names == 'Seoul'))
mask = (names == 'Seoul') | (names == 'chungju')
print(mask)
print(data[mask])
```

```python
문제 다음 데이터를 이용하여 두개 다 참일 때는 0, cond1이 참일 때는 1, cond2가 참일때는 2 그외일 때는 3으로 처리하여 출력
```

```python
result = []
cond1 = np.array([1,0,1,1,0,0,1], dtype = bool)
cond2 = np.array([0,1,1,1,0,1,1], dtype = bool)
print(cond1); print(cond2)
```

```python
for i in range(len(cond1)):
    if(cond1[i] and cond2[i]) :
        result.append(0)
    elif cond1[i] :
        result.append(1)
    elif cond2[i]:
        result.append(2)
    else :
        result.append(3)
print("조건문",result)
```

```python
result = np.where(cond1&cond2, 0, np.where(cond1, 1, np.where(cond2,2,3)))
print("3항 연산자",result)
```

```python
result = 1*(cond1& ~cond2) + 2*(cond2 & ~cond1) + 3 *~(cond1|cond2)
print("수식을 이용한", result)
```

# 행렬의 종류
- 정방행렬 : 정사각형, 행수와 열수가 같다.ex)공분산행렬, 상관행렬
- 대칭행렬 : 우상단과 좌하단 같은것
- 대각행렬 : scale을 변화시키는 요소
- 단위행렬(identity) : 행렬곱의 항등원, 행렬과 역행렬을 곱하면 단위행렬
- 역행렬 : 행렬의 나눗셈을 위해 구해서 곱해지는 행렬(정방행렬만 가능)
- 특이행렬 : 역행렬이 없는 행렬
- 직교행렬 : 축의 정직교, 각축간의 내적이 0인 행렬
- 정방행렬이면서 대칭행렬 : 공분산행렬, 상관계수행렬, 거리행렬
- 공분산 : 두 수간의 관계 ((한 수-평균) * (다른 수 - 평균))/(n-1)
- 상관계수(표준화 -1~1) : 공분산/표준편차의 곱

```python
import numpy as np
F1 = np.array([[1,2], [3,4]])
F2 = np.array([[2,2], [2,2]])
print(F1)
print(F2)
print(F1+F2) # 요소끼리
print(F1*F2)
print(F1@F2) #행렬곱도 내적, 두 벡터의 사이각을 구함
# 1,2    2,2,1
# 3,4    2,2,1    차원확장
```

```python
A = np.array([[2,1,-2], [3,0,1], [1,1,-1]])
print(A)
b = np.transpose(np.array([[-3,5,-2]])) # 열로 전치
print(b)
x = np.linalg.solve(A,b) # A의 역행렬을 구해서 b와 곱해줌
print(x)
```

```python
np.dot(np.linalg.inv(A),b) # 행렬곱
```

```python
a = np.array([0,1,0]) # y축
b = np.array([1,0,0]) # x축
res = np.dot(a,b) # 내적 사이각 = 직각
print(res)
```

```python
print(np.linalg.norm(a)) # 벡터의 크기
print(np.linalg.norm(b))
print(np.linalg.norm(a) * np.linalg.norm(b))
rad = res/(np.linalg.norm(a) * np.linalg.norm(b))
print("cos theta", rad)
print(np.arccos(rad)) # theta 값을 구함
print(np.degrees(np.arccos(rad))) # 사람이 알아보는 각도로 표현
```

```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
import pandas as pd

page = urlopen("https://news.naver.com/main/ranking/popularDay.nhn?rankingType=popular_day&sectionId=102&date=")
document = page.read()
page.close()
dt_index = pd.date_range(start='20200104', end = '20200204')
dt_index = dt_index.strftime("%Y%m%d").tolist()
dt_index
soup = BeautifulSoup(document, 'html5lib')
Title = []

title = soup.find('ol')
title_list = title.find_all("div", class_="ranking_headline")
# print(title_list)
for titles in title_list :
    Title.append(titles.get_text())
print(Title)
```

```python
import requests
import urllib.request
from bs4 import BeautifulSoup
import pandas as pd
import time
from datetime import date, time, datetime, timedelta

dt_index = pd.date_range(start='20200104', end = '20200204')
dt_index = dt_index.strftime("%Y%m%d").tolist()
dt_index
Title = []
for i in dt_index:
    url = 'https://news.naver.com/main/ranking/popularDay.nhn?rankingType=popular_day&sectionId=102&date='+i
    request = urllib.request.Request(url)
    response = urllib.request.urlopen(request)
    bs = BeautifulSoup(response,  'html.parser')
    title = bs.find('ol')
    title_list = title.find_all("div", class_="ranking_headline")
    for index in title_list:
        titles = index.get_text().strip()
        Title.append([titles])
    print(Title)
```

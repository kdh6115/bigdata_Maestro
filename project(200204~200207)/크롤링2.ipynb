```python
import requests
import urllib.request
from bs4 import BeautifulSoup
import pandas as pd
import time
from datetime import date, time, datetime, timedelta

dt_index = pd.date_range(start='20190204', end = '20200204')
dt_index = dt_index.strftime("%Y%m%d").tolist()
dt_index
society = []
for i in dt_index:
    url = 'https://news.naver.com/main/ranking/popularDay.nhn?rankingType=popular_day&sectionId=102&date='+i
    request = urllib.request.Request(url)
    response = urllib.request.urlopen(request)
    bs = BeautifulSoup(response,  'html.parser')
    bs_class = bs.select('.ranking_headline')
    for index in bs_class:
        title = index.get_text().strip()
        society.append([title])
    print(i)
```

```python
df = pd.DataFrame.from_records(society)
df.to_excel('society.xlsx')
```
